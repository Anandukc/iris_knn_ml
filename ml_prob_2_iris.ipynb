```python
df=df.drop(['Id'],axis=1)
df
```

```python
# seperate x and y
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
# convert to training and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
x_train
```

```python
x_test
```

```python
from pandas.core.frame import Scalar
# normalisation:
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.fit_transform(x_train) #normalisation
x_test=scaler.fit_transform(x_test)    #normalisation
x_train
```

```python
x_test
```

```python
# implement model
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=9)
model.fit(x_train,y_train)
```

```python
#predict
y_predict=model.predict(x_test)
y_predict
```

```python
# predict using user data
pred=model.predict(scaler.transform([[6.4,3.2,5.3,2]]))
pred
```

```python
# evaluation
from sklearn.metrics import confusion_matrix,accuracy_score,ConfusionMatrixDisplay,classification_report
result=confusion_matrix(y_test,y_predict)
labels=['Iris-setosa','Iris-versicolor','Iris-virginica']
cmd=ConfusionMatrixDisplay(result,display_labels=labels)
cmd.plot()
score=accuracy_score(y_test,y_predict)


```
